## How to use loc and iloc

```python
df = pd.DataFrame(np.arange(20).reshape((5,4)),
                  index=["1st","2nd","3rd","4th","5th"],
                  columns=["a","b","c","d"])
df
```

loc: extract data by index name and column name

```python
df.loc[:,"a"]
```

```python
df.loc[:, ["a","c"]]  #a ,c baganuud
```

```python
df.loc["3rd", :]  #3rd row
```

```python
df.loc[["2nd", "4th"], :] #2nd, 4th row
```

```python
df.loc[["2nd","5th"],["b","d"]]   #2nd, 5th rowiin b,d bagana
```

iloc: extract by index number, column number

```python
df.iloc[0,2]  #column number 2 with index number 0
```

```python
df.iloc[2:,3]
```

```python
df.iloc[1:, :2]
```

```python
df.tail(2)  #suuliin muruud
```
