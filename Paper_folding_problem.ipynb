## [Problem 1] Implementation using exponentiation arithmetic operators

```python
#method1
THICKNESS = 0.00008

folded_thickness = THICKNESS * pow(2,43)

print("Thickness: {} meters".format(folded_thickness))
```

## [Problem 2] Unit Conversion

```python
#method1
THICKNESS = 0.00008

folded_thickness = THICKNESS * pow(2,43)

print("Thickness: {: .2f} kilometers".format(folded_thickness / 1000))
```

## [Problem 3] Create using a for statement

```python
#method2
THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(43):
  folded_thickness = folded_thickness * 2
print("Thickness: {: .2f} kilometers".format(folded_thickness / 1000))
```

## [Problem 4] Comparison of calculation time

Both of the above two methods are correct, but when comparing the goodness of the code, for example, the following points are evaluated.

Execution speed
Memory usage
Readability
Scalability
Reusability

```python
#method1
import time
start = time.time()

THICKNESS = 0.00008

folded_thickness = THICKNESS * pow(2,43)

print("Thickness: {: .2f} kilometers".format(folded_thickness / 1000))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
#method2
import time
start = time.time()

THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(43):
  folded_thickness = folded_thickness * 2
print("Thickness: {: .2f} kilometers".format(folded_thickness / 1000))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

Speed: method1 > method2

```python
#method1
%%timeit
import time
start = time.time()

THICKNESS = 0.00008

folded_thickness = THICKNESS * pow(2,43)

print("Thickness: {: .2f} kilometers".format(folded_thickness / 1000))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
#method2
%%timeit
import time
start = time.time()

THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(43):
  folded_thickness = folded_thickness * 2
print("Thickness: {: .2f} kilometers".format(folded_thickness / 1000))

elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

## [Problem 5] Saving to a list

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS
process = []
process.append(folded_thickness)
for i in range(43):
  folded_thickness = folded_thickness * 2
  process.append(folded_thickness)
print("Thickness: {: .2f} kilometers".format(folded_thickness / 1000))
print(process)
print(len(process))
```

## [Problem 6] Displaying a line graph

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Display the graph. Title and axis label name.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(process)  # Enter the variable name of the list in "List name"
plt.show()
```
