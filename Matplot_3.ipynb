## Drawing pie chart

```python
import pandas as pd
df = pd.DataFrame([['Advertising on TV, radio, newspapers, etc', 5],
                   ['internet search results', 15],
                   ['advertising on the internet', 25],
                   ['SNS advertisement', 50],
                   ['Referral from an acquaintance', 1],
                   ['others', 4]], columns = ["media", "Aggregate results（%）"])
df
```

```python
labels = df.loc[:,"media"]
rate = df.loc[:, "Aggregate results（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels=labels)
ax.axis('equal')
plt.show()
```

## Pie chart formatting

By default, elements counterclockwise (3 o'clock)


change starting position

```python
labels = df.loc[:,"media"]
rate = df.loc[:, "Aggregate results（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels = labels, startangle=90)
ax.axis('equal')
plt.show()
```

```python
labels = df.loc[:,"media"]
rate = df.loc[:, "Aggregate results（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels = labels, startangle=90, counterclock=False)
ax.axis('equal')
plt.show()
```

Add shadow to pie chart

```python
labels = df.loc[:,"media"]
rate = df.loc[:, "Aggregate results（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels = labels, startangle=90, counterclock=False, shadow=True)
ax.axis('equal')
plt.show()
```

express percentage of each elements

```python
labels = df.loc[:,"media"]
rate = df.loc[:, "Aggregate results（%）"]

fig, ax = plt.subplots()
ax.pie(rate, labels = labels, startangle=90,counterclock=False, shadow=True, autopct='%1.0f%%')
ax.axis('equal')
plt.show()
```

explode ashiglasnaar specific element

```python
labels = df.loc[:,"media"]
rate = df.loc[:, "Aggregate results（%）"]

fig, ax = plt.subplots()
explode=[0.2, 0, 0, 0, 0, 0.2]
ax.pie(rate, explode=explode, labels = labels, startangle=90,counterclock=False, shadow=True, autopct='%1.0f%%')
ax.axis('equal')
plt.show()
```

# Multiple graph drawing and dataframe utilization

## Draw multiple graphs

```python
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4]
y1 = [5, 4, 7, 8]
x2 = [1, 2, 3, 4]
y2 = [4, 3, 6, 7]

fig, ax = plt.subplots()
ax.plot(x1, y1, label='data1')
ax.bar(x2, y2, label='data2')
ax.legend()
plt.show()
```

## Drawing a DataFrame

line graph

```python
import pandas as pd
df = pd.DataFrame({'data1': [3, 2, 1], 'data2': [1, 4, 2]})
import matplotlib.pyplot as plt 
df.plot()
plt.show()
```

bar graph

```python
df = pd.DataFrame({'data1': [3, 2, 1], 'data2': [1, 4, 2]})
df.plot.bar()
plt.show()
```

stacked bar chart

```python
df = pd.DataFrame({'data1': [3, 2, 1], 'data2': [1, 4, 2]})
df.plot.bar(stacked=True)
plt.show()
```

## Advanced formatting

graph color

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, color='red')
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.bar(x, y, color='white', edgecolor='red')
plt.show()
```

Line style

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=10, color='g')
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, linestyle='--', linewidth=5, color='k')
plt.show()
```

```python
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [6, 4, 2]

fig, ax = plt.subplots()
ax.plot(x1, y1, linestyle='-.')
ax.plot(x2, y2, linestyle=':')
plt.show()
```

Font style

```python
fig, ax = plt.subplots()
ax.set_title('Edit font style')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title('Edit font style', family='monospace', size=20, weight='bold')
plt.show()
```

font dahin ashiglah

```python
fontdict = {
  'family': 'monospace',
  'size': 20,
  'weight': 'bold'
}

fig, ax = plt.subplots()
ax.set_title('Graph title', fontdict=fontdict, size=30)
ax.set_xlabel('x label', fontdict=fontdict)
ax.set_ylabel('y label', fontdict=fontdict)
plt.show()
```
