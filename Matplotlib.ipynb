### Title

suptitle, set_title function

```python
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [-2, -4, -6]

fig, axes = plt.subplots(1,2)
axes[0].plot(x1,y1)
axes[1].plot(x2,y2)
axes[0].set_title('Subplot1 title')
axes[1].set_title('Subplot2 title')
fig.suptitle('Graph title')
plt.show()
```

#### axis label

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
plt.show()
```

#### usage guide

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, label='sample legend')
ax.legend()
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

fig, ax = plt.subplots()
ax.plot(x, y, label = 'sample legend')
ax.legend(loc='center left')
plt.show()
```

# Bar graph

## draw bar chart

```python
x = [1, 2, 3]
y = [8, 2, 4]

fig, ax = plt.subplots()
ax.bar(x, y)
plt.show()
```

### draw graph horizontally

```python
x = [1, 2, 3]
y = [8, 2, 4]

labels = ['label1', 'label2', 'label3']

fig, ax = plt.subplots()
ax.bar(x, y, tick_label=labels, color = 'green')
plt.show()
```

```python
x = [1, 2, 3]
y = [8, 2, 4]

labels = ['label1', 'label2', 'label3']

fig, ax = plt.subplots()
ax.barh(x, y, tick_label=labels, color = 'green')
plt.show()
```

### Arrange multiple bar charts

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]

ax.bar(x, y1, label='y1', color='yellow')
ax.bar(x, y2, label='y2', color='blue')
ax.legend()
plt.show()
```

2 dah baruudiig shift hiih width tohiruulah

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
width = 0.4
ax.bar(x, y1, label='y1', width=width)
ax.bar(x, y2, label='y2', width=width)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
width = 0.4

ax.bar(x, y1, width=width, label='y1')

x2 = [num + width for num in x]

ax.bar(x2, y2, width=width, label='y2')
ax.legend()
plt.show()
```

xticks, shifted from the center

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
labels = ['Apple', 'Banana', 'Orange']
width = 0.4

ax.bar(x, y1, width=width, label='y1')
x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='y2')
ax.legend()
plt.xticks(x, labels)
plt.show()
```

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
labels = ['Apple', 'Banana', 'Orange']
width = 0.4

ax.bar(x, y1, width=width, label='y1')

x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='y2')
ax.legend()

label_axis = [num + width/2 for num in x]
plt.xticks(label_axis, labels)
plt.show()
```

# Stacked bar charts

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
y_label = np.add(y1, y2)
labels = ['Apple', 'Banana', 'Orange']
ax.bar(x, y_total, tick_label=labels, label='y2')
ax.bar(x, y1, label='y1')
ax.legend()
plt.show()
```

3 data

```python
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [4, 12, 5]
y2 = [8, 2, 4]
y3 = [2, 6, 8]
y1_y2_total = np.add(y1, y2)
y_total = np.add(y1_y2_total, y3)
labels = ['Apple', 'Banana', 'Orange']
ax.bar(x, y_total, tick_label=labels, label='y3')
ax.bar(x, y1_y2_total, tick_label=labels, label='y2')
ax.bar(x, y1, label='y1')
ax.legend()
plt.show()
```
