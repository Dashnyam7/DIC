# Different operation on pandas

## Read file data

```python
import numpy as np
import pandas as pd
df = pd.read_csv("sales_data.csv")
df
```

## Data type conversion

```python
df.dtypes
```

```python
df.loc[:, '日付'] = df.loc[:, '日付'].apply(pd.to_datetime)
```

```python
df.dtypes   #pandas sangiin apply function
```

```python
df.loc[:, '客単価'] = df.loc[:, '客単価'].astype(np.float32)
df    #numpy sangiin astype
```

## Sort data

```python
df.sort_values(by="売り上げ")
```

low-high: ascending order,  
high-low: descending order

```python
df.sort_values(by="売り上げ", ascending=False)
```

## Delete column

```python
df = df.drop("客単価", axis=1)
df
```

## Calculation bewteen columns

```python
df.loc[:, "客単価"] = df.loc[:, "売り上げ"]/df.loc[:, "客数"]
df
```

```python
df.loc[:, "客単価"] = df.loc[:, "客単価"].round().astype(int)
df
```

# Handle time series data and missing values

## Creating time series data

```python
import numpy as np
import pandas as pd
dates = pd.date_range(start="2022-10-19", end="2022-12-31")
dates
```

Create dataframe

```python
df = pd.DataFrame(np.random.randint(1,100, 74), index=dates, columns=["udur"])
df
```

## Data grouping

Sar buriin dundaj utga

```python
df.groupby(pd.Grouper(freq='M')).mean() 
```

```python
df.groupby(pd.Grouper(freq='W')).mean() 
```

```python
df.resample('M').mean()
```

## Missing value imputation

```python
df = pd.read_csv("sales_data_contains_nan.csv")
df
```

Remove rows with missing values

```python
dropna_df = df.dropna()
dropna_df
```

impute missing values with specified value

```python
fillna_df = df.fillna(0)
fillna_df   #nan hoosong 0-uudaar duurgene
```

```python
ffill_df = df.fillna(method = 'ffill')
ffill_df    #hooson zai ni deed taliin utgatai adil
```

Imputation with mean, median and mode

```python
fill_mean_df = df.fillna(df.mean())
fill_mean_df
```

```python
df.mode()
```

```python
fill_freq_df = df.fillna(df.mode().iloc[0, :])
fill_freq_df
```
